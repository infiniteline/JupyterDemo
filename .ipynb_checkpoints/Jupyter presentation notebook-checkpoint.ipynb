# Plotting

```python
from pandas.plotting import scatter_matrix
%matplotlib inline
%config InlineBackend.figure_format = 'retina'

arousalDF = testDF.loc[:, testDF.columns.str.contains('Arousal')]
scatter_matrix(arousalDF, alpha=0.2, figsize=(10, 10), diagonal='kde');
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

corr = testDF.corr()
fig, ax = plt.subplots(figsize=(10,10))
sns.heatmap(corr, xticklabels=corr.columns.values, yticklabels=corr.columns.values, annot=True, square=True, cmap="YlGnBu");
```

```python
valenceAllDF = testDF.loc[:, ["Image","Valence Overall","Valence Female","Valence Male"]]
valenceAllDF = valenceAllDF.sort_values('Image')
valenceAllDF = valenceAllDF.set_index('Image')
pd.set_option('display.max_rows', 100)
valenceAllDF
```

```python
from bokeh.io import output_notebook
from bokeh.models import (
    ColumnDataSource,
    HoverTool,
    LinearColorMapper,
    BasicTicker,
    PrintfTickFormatter,
    ColorBar,
)
from bokeh.plotting import figure, show

output_notebook()

TOOLS = "hover,save,pan,box_zoom,reset,wheel_zoom"
colors = ["#55766f", "#6f9990", "#c9d9d3", "#e2e2e2", "#dfccce", "#b2656a", "#933b41"]
mapper = LinearColorMapper(palette=colors, low=0, high=7)

images = list(valenceAllDF.index)
categories = list(valenceAllDF.columns)

plotSourceDF = pd.DataFrame(valenceAllDF.stack(), columns=['Valence']).reset_index()
plotSourceDF.columns = ["Image", "Category", "Valence"]

p = figure(title="OASIS Images Valense Response by Gender",
           x_axis_location="above", plot_width=900, plot_height=400, x_range=images, y_range=categories,
           tools=TOOLS, toolbar_location='below')

p.grid.grid_line_color = None
p.axis.axis_line_color = None
p.axis.major_tick_line_color = None
p.axis.major_label_text_font_size = "8pt"
p.axis.major_label_standoff = 0
p.xaxis.major_label_orientation = 3.14 / 3

p.rect(x="Image", y="Category", width=1, height=1,
       source=ColumnDataSource(plotSourceDF),
       fill_color={'field': 'Valence', 'transform': mapper},
       line_color=None)

color_bar = ColorBar(color_mapper=mapper, major_label_text_font_size="5pt",
                     ticker=BasicTicker(desired_num_ticks=len(colors)),
                     formatter=PrintfTickFormatter(format="%d"),
                     label_standoff=6, border_line_color=None, location=(0, 0))
p.add_layout(color_bar, 'right')

p.select_one(HoverTool).tooltips = [
     ('Image', '@Image'),
     ('Valence', '@Valence'),
]

show(p)
```
